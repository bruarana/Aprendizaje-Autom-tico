```python
#Ejercicio 1: Array de 10 ceros
import numpy as np

arr1 = np.zeros(10)
print(arr1)
```

```python
#Ejercicio 2: Array del 1 al 20
arr2 = np.arange(1, 21)
print(arr2)
```

```python
#Ejercicio 3: Números pares del 2 al 50
arr3 = np.arange(2, 51, 2)
print(arr3)
```

```python
#Ejercicio 5: Matriz 4x4 con valores aleatorios (0 a 1)
mat1 = np.random.rand(4, 4)
print(mat1)
```

```python
#Ejercicio 6: Suma elemento a elemento
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

suma = a + b
print(suma)
```

```python
#Ejercicio 7: Multiplicación de matrices 3x3
mat2 = np.random.randint(1, 11, (3, 3))
mat3 = np.random.randint(1, 11, (3, 3))

producto = np.dot(mat2, mat3)

print("Matriz A:\n", mat2)
print("Matriz B:\n", mat3)
print("Producto:\n", producto)
```

```python
#Ejercicio 8: Array aleatorio (1 a 100) + estadísticas
arr4 = np.random.randint(1, 101, 15)

print("Array:", arr4)
print("Máximo:", np.max(arr4))
print("Mínimo:", np.min(arr4))
print("Promedio:", np.mean(arr4))
```

```python
#Ejercicio 9: Matriz 5x5 (1 al 25) y diagonal
mat4 = np.arange(1, 26).reshape(5, 5)

diagonal = np.diag(mat4)

print(mat4)
print("Diagonal:", diagonal)
```

```python
#Ejercicio 10: Suma por filas y columnas
mat5 = np.random.randint(1, 21, (5, 5))

suma_filas = np.sum(mat5, axis=1)
suma_columnas = np.sum(mat5, axis=0)

print("Matriz:\n", mat5)
print("Suma filas:", suma_filas)
print("Suma columnas:", suma_columnas)
```
